```python
import pandas as pd
dataset = pd.read_csv('50_Startups.csv')
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 
       'State_Florida', 'State_New York']]
independent
```

```python
dependent = dataset[['Profit']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor( criterion='friedman_mse', splitter='best',)
regressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_pred,y_test)
r_score
```

```python
import pickle
filename = 'finalized_DecisionTree_model.sav'
pickle.dump(regressor,open(filename,'wb'))
```

```python
load_model = pickle.load(open('finalized_DecisionTree_model.sav','rb'))
result = load_model.predict([[300,400,300,1,0]])
```
